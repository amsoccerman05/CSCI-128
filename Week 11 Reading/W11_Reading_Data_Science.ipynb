## Introduction

### What is Data Science?

Data science is an interdisciplinary field focused on extracting meaningful insights from data. It uses techniques from Mathematics, Statistics, and Computer Science to develop methods to  **utilize**, **explore**, and **visualize** data. Domain knowledge can then be applied to analyze this data and create value from it.

Data Science is one of the most widely applicable uses of programming. Whatever you choose for your future career, it will likely require you to produce, handle, and interpret digital data. There are several tools to make this process possible without programming knowledge, such as Microsoft Excel, however, programming will make the process significantly faster, more flexible, and easier. Additionally, programming enables you to work with large datasets and create reproducible processes to automate the work you do in the future.



### The Data Science Process

The Data Science process consists of four steps: obtaining data, cleaning data, exploring the data, and finally predicting unknowns. Let's take a closer look at each of these processes:

#### 1. Obtain Data
Obtaining data can either be the easiest or hardest part of the process. Many institutions provide open data for transparency and research purposes. There are also many websites where you can obtain free datasets. For example, [Kaggle](https://www.kaggle.com/) provides hundreds of datasets for teaching and practicing data science. Meanwhile [Dataset Search](https://datasetsearch.research.google.com/) from Google can help you find specific datasets.

However, there are also many datasets that are quite hard (sometimes impossible) to obtain. For instance, sales and marketing data of products are highly valuable for the companies that produce them, that's why they pay notable amounts of money to market research companies to gather that information. Furthermore, giant retailers like Walmart and Amazon naturally generate these dataset from their own sales records and this gives them the chance to optimize their supply chains, store/website layouts and generate better sales strategies. If you read their stories, you will see that they both owe their curent success to data science!

#### 2. Clean Data

As digital data are generated by multiple sources, values in them usually need several steps of preprocessing to be standardized. Manual entry of data is still a significant source of digital data, and creates many errors. Therefore, we need to clean the data before we can explore it. 

Usually this step is the most time consuming and boring part of data science. However, it's extremely crucial! Otherwise, your data may generate meaningless results, or even worse, may lead you to make bad decisions.

#### 3. Explore Data

Exploring the data is essentially the act of finding key insights and patterns that the data provides. To be able to find these insights, it is helpful to visualize the data in several ways, apply statistical methods, and find clues that lead us to these hidden insights like a detective.

#### 4. Predict Unknowns

This step is where we create value from the insights we find in the data. Usually, this step requires domain knowledge where you may capitalize on the insight.

For instance, you might find cheese and macaroni are frequently bought together. If you know the domain, you know it's because people love mac and cheese. You can create value from this insight by recommending a store owner place them closer to each other to increase their sales or create a new product where you sell both together. On the other hand, if you don't know the domain very well, you may conclude that the lack of pirates is the cause of global warming, like the following visual suggests:

### 1- Obtaining the data

Our first dataset is Nutrition data from selected cereal products. It is adapted from Kaggle. 

First, we need to read the file. Our file is in .csv format. This format consists of a series of values separated by commas, it can also be referred to as a **Comma Separated Value** file. Note that .csv files are still just text files, we can interact with it the same way we learned last week, but knowing it is also a csv tells us something valuable about the structure of the data it contains. This makes it easier to parse and interact with that data.

To interact with text files as csv's, we need to import the Python csv module and use it to create reader or writer objects. The reason we are interested in doing this is because the reader will automatically detect and remove the commas separating values for us. Essentially, the reader will split the data up into lists for us. This makes it simpler to navigate and interact with the data in the file. The example below shows one way to do this, but we also recommend taking a look at the [official documentation](https://docs.python.org/3/library/csv.html) for the csv module.

For our Python program, we will create a list and store the lines of the file in that list.

```python
import csv
cereal_data = []
with open('cereal.csv', newline='') as csvfile:
    cereals = csv.reader(csvfile,delimiter=',')
    for row in cereals:
        cereal_data.append(row)


cereal_data
```

As we may detect from the output, it is a 2D list! Datasets are commonly in forms of 2D lists (also often called tables or matrices), where each line represents a data point and each column represent an attribute. Examining the program, you can see we only had to write a few extra lines to make use of the csv module. The result is that each line of the file is automatically split and presented to us as a list.

```python
print(f'Number of rows (data points): {len(cereal_data)}')
print(f'Number of columns (variables): {len(cereal_data[0])}')
```

We have 78 rows and 6 columns. However, there is something else we missed. It is the header row, which provides us the meaning of the columns (attributes).

```python
print(cereal_data[0])
```

As you can see, the first line is the header. It doesn't contain actual data, but it does tell us a lot of useful information about the format the rest of the data file is in. You may notice not all of the headers are clear. Let's fix that!

```python
cereal_data[0][1]='manufacturer'
print(cereal_data[0])
```

Now let's check the content under the header. The manufacturers are provided with their initials:

- A = American Home Food Products
- G = General Mills
- K = Kelloggs
- N = Nabisco
- P = Post
- Q = Quaker Oats
- R = Ralston Purina

'calories', 'protein', and 'sugars' are the corresponding values per serving and 'rating' is the average consumer review of the product.

We are going to perform statistical analysis on these columns, but first we need a function to obtain the columns from our dataset.

```python
def col_func(matrix, i):
    column = []
    for row in matrix:
        column.append(row[i])
    column.pop(0)
    return column

print(col_func(cereal_data,3))
```

We can now get any columns as lists for our use. Let's start analyzing some of our data!

### 2- Statistical Analysis

We will now perform statistical analysis on our columns. Our primary goal is to try to understand more about the data we have. A nice starting point is to run five common descriptive statistics on each column. This will help us understand more about our variables:

- Min: Minimum value of the given column
- Max: Maximum value of the given column
- Mean: Average of the given column
- Median: Median is the value in the middle when we sort all the values in the given column
- Mode: The value that appears the most in the given column

Let's start with min and max. Fortunately, they are built-in functions in Python.



```python
for i in range(2,6):
    
    corresponding_col = col_func(cereal_data,i)
    
    for j in range(len(corresponding_col)):
        corresponding_col[j] = float(corresponding_col[j])
    
    min_value = min(corresponding_col)
    max_value = max(corresponding_col)
    
    print(f'Min of {cereal_data[0][i]} is {min_value}')
    print(f'Max of {cereal_data[0][i]} is {max_value}')
```

It looks mostly good, but wait a minute...

There must be a problem! Sugar and rating cannot be negative and rating should be below 100. This is because we forget a step: the data cleaning. To be able to move on we have to clean the data. Obviously incorrect values, frequently called "outliers", are a common data quality issue. There are several others such as duplicate points, redundant columns, and missing cells. For simplicity, in our case, incorrect values are the only issue. 

We cannot know the real values for these cells, therefore we will get rid of these rows (data points) completely. 


```python
cereal_data_clean =[]
cereal_data_clean.append(cereal_data[0])
for row in cereal_data[1:]:
    if float(row[2]) < 0 or float(row[3]) < 0 or float(row[4]) < 0 or float(row[5]) < 0 or float(row[5]) > 100:
        continue
    else:
        cereal_data_clean.append(row)

print(f'Number of rows (data points): {len(cereal_data_clean)}')
```

After cleaning the data we only lost 5 data points out of 78. While working with data, we usually want to have as many data points as we can, but in order to assure the quality of the analysis we sometimes have to sacrifice them.

Let's move back to statistical analysis.

```python
for i in range(2,6):
    
    corresponding_col = col_func(cereal_data_clean,i)
    
    for j in range(len(corresponding_col)):
        corresponding_col[j] = float(corresponding_col[j])
    
    min_value = min(corresponding_col)
    max_value = max(corresponding_col)
    
    print(f'Min of {cereal_data[0][i]} is {min_value}')
    print(f'Max of {cereal_data[0][i]} is {max_value}')
```

Looking good! Let's calculate the remaining descriptive statistics. Many python packages already have these functions implemented, however it is good practice to write down these functions at least once to understand how they work:



```python
def Mean(column):
    sum_val = 0
    for i in range(len(column)):
        sum_val = sum_val + float(column[i])
    
    return sum_val/len(column)

def Median(column):
    for i in range(len(column)):
        column[i]=float(column[i])
    
    column.sort()
    if(len(column)%2==0):
        median = (column[len(column)//2]+ column[(len(column)//2)-1])/2
        return median
    else:
        median = column[len(column//2)]
        return median

for i in range(2,6):
    
    corresponding_col = col_func(cereal_data_clean,i)
    
    print(f'Mean of {cereal_data[0][i]} is {Mean(corresponding_col)}')
    print(f'Median of {cereal_data[0][i]} is {Median(corresponding_col)}')
```

Great! This is quite important information to understand the structure of our dataset. However, it might not be enough to represent everything about our data. [Anscombe's Quartet](https://en.wikipedia.org/wiki/Anscombe%27s_quartet) is a great example of why descriptive statistics may sometimes mislead us. Consequently, we should also visualize our data to be able to comprehend its details better.

### 3- Visualize!

To visualize our dataset we will use a package called pyplot, part of a larger module called matplotlib. Matplotlib is a very popular, very robust visualization package, and pyplot is a simplified version of it. Let's start with our first visualization. 

#### Histogram

A histogram is used to understand the distribution of one column (variable) of data. You'll see we can generate a histogram very quickly with pyplot.

```python
col_to_visualize = col_func(cereal_data_clean,2)
for i in range(len(col_to_visualize)):
        col_to_visualize[i]=float(col_to_visualize[i])


plt.hist(col_to_visualize) # This is the function to create our histogram object
plt.show()  # This function is used after all our other pyplot functions to show the final results
```

Good, but it is hard to understand what we are observing. It would be clearer if we had some labels in our visualization.

```python
col_to_visualize = col_func(cereal_data_clean,2)
for i in range(len(col_to_visualize)):
        col_to_visualize[i]=float(col_to_visualize[i])


plt.hist(col_to_visualize)
plt.title(f'Histogram of {cereal_data_clean[0][2]}')
plt.xlabel(f'{cereal_data_clean[0][2]}')
plt.ylabel('Number of Observations')
plt.show()
```

We can observe that 3 cereals have less than 60 calories, 2 have between 60 and 65, and so on and so forth. The majority of the observations fall into bins in the middle. The majority of cereals have between 90 and 120 calories per serving. This is valuable information to know, and much easier to learn via a histogram than descriptive statistics.

You may remember there are many different companies in our list. Let's compare the calorie content of their products against each other. First, we need to filter the dataset by companies and then calculate the mean of the calorie values of cereals that those companies produce. 

```python
def mean_filter(data,company,col_index):
    sum_val = 0
    count = 0
    for row in data:
        if row[1] == company:
            count = count+1
            sum_val = sum_val + float(row[col_index])
    return sum_val / count

mean_val = mean_filter(cereal_data_clean,'A',2)
print(f'Average calories in General Mills products: {mean_val}' )

```

Let's compare all these companies with each other:



```python
companies= ['A','G','K','N','P','Q','R']
means = []
for company in companies:
    means.append(mean_filter(cereal_data_clean,company,2)) 

for i in range(len(companies)):
    print(f'Average Calorie of Cereals from {companies[i]} is {means[i]}')
```

It's not so easy to compare, is it? Maybe we can find an appropiate visualization for this kind of situation too!

Note: Remember the companies
- A = American Home Food Products
- G = General Mills
- K = Kelloggs
- N = Nabisco
- P = Post

- R = Ralston Purina



#### Bar Charts

To compare multiple values, we use side by side bars called bar charts.

```python
plt.bar(companies,means)
plt.title(f'Comparison of average calories')
plt.xlabel(f'Companies')
plt.ylabel('Calories')
plt.show()
```

Now it's obvious that cereals of Nabisco contain the least amount of calories on average, while Ralstons Purina has the most. 

It's well known that calories can be associated with amount of sugar. Let's see if we can verify this is also true for this dataset!

#### Scatter Plots

Scatter Plots show the relationship between two numeric variables

```python
calories = col_func(cereal_data_clean,2)
sugars = col_func(cereal_data_clean,4)

for i in range(len(calories)):
        calories[i]=float(calories[i])
        sugars[i]=float(sugars[i])


plt.scatter(calories,sugars)
plt.title(f'Comparison of sugars and calories')
plt.xlabel(f'Calories per serving')
plt.ylabel('Sugars per serving (g)')
plt.show()
```

We can see that even though they do not exactly follow each other, there is a relationship between sugars and calories. This is because the data points are somewhat "grouped" from bottom left to top right. If sugar amount always predicted calorie amount exactly, then all points would fall on a single, positively-sloped line. If there was no relationship between sugar and calories, then the data points would look like they had all been randomly placed across the plot.

Finally, we will work on one more type of plot, a line plot.

### Bias

As we analyze our data to shed light on the unknown, some mistakes can create different biases in our results. These can be **Confirmation bias**, which is assuming our knowledge prior to the analysis is correct and trying to adapt the results of our analysis based on those beliefs; **Selection Bias**, which is using a dataset that is not a good representation of the general problem; or **Historical Bias**, which is using historical data that contains inequalities or prejudices to influence decisions in the future. [This Article](https://www.metabase.com/blog/6-most-common-type-of-data-bias-in-data-analysis) provides extensive information on 6 of the most common types of data biases.
